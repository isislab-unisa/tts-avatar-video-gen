```python
from IPython.display import Audio
from TTS.api import TTS
from torch.serialization import add_safe_globals
from TTS.tts.configs.xtts_config import XttsConfig
from TTS.tts.models.xtts import XttsAudioConfig, XttsArgs
from TTS.config.shared_configs import BaseDatasetConfig
import torch
print(torch.__file__)
print(torch.__version__)
add_safe_globals([XttsConfig])
add_safe_globals([XttsAudioConfig])
add_safe_globals([BaseDatasetConfig])
add_safe_globals([XttsArgs])

# Get device
device = "cuda" if torch.cuda.is_available() else "cpu"

# List available 🐸TTS models
print(TTS().list_models())

model = "tts_models/multilingual/multi-dataset/xtts_v2"

# Initialize TTS
tts = TTS(model).to(device)

# List speakers
print(tts.speakers)

# Run TTS
# ❗ XTTS supports both, but many models allow only one of the `speaker` and
# `speaker_wav` arguments

# TTS with list of amplitude values as output, clone the voice from `speaker_wav`
# reference = "reference/it_speaker_9.mp3"
reference = "reference/en_speaker_9.mp3"

wav = tts.tts(
  text="Salve sono Cody, e sono qui per presentarvi il corso di studio di informatica presso l'università di Fisciano",
  speaker_wav=reference,
  # speaker="Eugenio Mataracı",
  language="it",
  split_sentences=True,
)

# TTS to a file, use a preset speaker
# tts.tts_to_file(
#   text="Salve sono Cody, e sono qui per presentarvi il corso di studio di informatica presso l'università di Fisciano",
#   # speaker="Craig Gutsy",
#   # speaker_wav=reference,
#   # language="it",
#   file_path="output.wav"
# )

#Plat the audio using Audio
print("AUDIO RATE:", tts.synthesizer.output_sample_rate)

Audio(wav, rate=tts.synthesizer.output_sample_rate)
```
